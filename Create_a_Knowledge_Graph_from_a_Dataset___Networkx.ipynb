```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/movies.csv')
```

```python
df.head(10)
```

```python
titles = df['title'].unique().tolist()
titles
```

```python
genres = df['genres'].unique().tolist()
genres
```

```python
nodes = titles + genres
nodes
```

```python
edges = []
for index, row in df.iterrows():
  title = row['title']
  genres = row['genres'].split('|')
  for genre in genres:
    edges.append((title, genre))
```

```python
graph = nx.Graph()
graph.add_nodes_from(nodes)
graph.add_edges_from(edges)
```
